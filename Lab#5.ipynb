# Reading Data
The pickle module implements binary protocols for serializing and de-serializing a Python object structure. The data format used by pickle is Python-specific. This has the advantage that there are no restrictions imposed by external standards such as JSON or XDR (which canâ€™t represent pointer sharing); however it means that non-Python programs may not be able to reconstruct pickled Python objects.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import pickle
import pandas as pd
```

```python
# reading data

infile = open ("higgs_100000_pt_1000_1200.pkl",'rb')
higgs_data = pickle.load(infile)

infile2 = open ("qcd_100000_pt_1000_1200.pkl",'rb')
qcd_data = pickle.load(infile2)
```

Accroding to the lab instruction we will be looking for the standard model of the Higgs Boson, in a protons collision at the center of mass energy of 13 TeV, which is at the Large Hadron Collider(LHC). Higgs Boson is basically the reason why we know subatomic particles have masses, an elementary particle in the Standard Model of particle physics produced by the quantum excitation of the Higgs field, one of the fields in particle physics theory. And the Large Hadron Collider is the world's largest and highest-energy particle collider. It's able to make protons smash into each other and by doing that scientists are able to find what protons are actually made of. Smashing protons creates quantum excitation in the higgs field, which makes flicking off a higgs particle possible. However the odds of that happening is extremly rare and it happens about once in every trillion collisions.

```python
# now we look at the keys in our dataset
print("Higgs Signal categories: ", higgs_data.keys())
print("QCD background categories: ", qcd_data.keys())
```

What are these dataset variables?

pt - Corresponds to the transverse momentum in the xy-plane of the proton to proton collisions in the LHC.\
eta(n) - or theta is the pseudorapidity, or the angle of a particle relative to the beam axis. The angle of the particle in the z-y plane.\
phi - The azimuthal angle responsible for the angle of the particle in x-y plane.\
mass - invariant mass, or the intrinsic mass of the system of objects independent of the overall motion of the system.\
ee2(e2) - 2-point ECF ratio, is the energy correlation function ratio for 2-point correlators. Particularly useful for quark/gluon discrimination.\
ee3(e3) - 3-point ECF ratio, is the energy correlation function ratio for 3-point correlators. Particularly useful for boosted W/Z/Higgs boson identification.\
d2 - 3 to 2 point ECF ratio\
angularity(tau) - Specific orientation of one feature to another at a referenced angle.\
Jet - signatures of quarks,gluons,etc. created in high-energy collisions.\
N- subjettiness - a jet shape designed to identify boosted hadronic objects such as top quarks. Given N subjet axes within a jet, N-subjettiness sums the angular distances of jet constituents to their nearest subjet axis.\
t1 - 1-subjettiness\
t2 - 2-subjettiness\
t3 - 3-subjettiness\
t21 - 21-subjettiness\signal = np.array(higgs_data)
qcd = np.array(qcd_data)
t32 - 32-subjettiness\
KtDeltaR - delta R of two subjets within the large-R jet

```python
signal = np.array(higgs_data)
qcd = np.array(qcd_data)
```

First we make histograms of each field for each feature in both the background and signal datasets.

```python
fig, (ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10,ax11,ax12,ax13,ax14) = plt.subplots(14,1)
axes = (ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10,ax11,ax12,ax13,ax14)
labels = ['pt', 'eta', 'phi', 'mass', 'ee2', 'ee3', 'd2', 'angularity', 't1', 't2', 't3', 't21', 't32', 'KtDeltaR']
axes[0].set_title("Histogram of all the Parameters")

for i in range(0,14):
    axes[i].hist(signal[:,i],bins = 50, density = True, alpha = 0.7, label="Signal")
    axes[i].hist(qcd[:,i],bins = 50, density = True, alpha = 0.7, label = "Background")
    axes[i].set_xlabel(labels[i])
    axes[i].legend(loc = 'best')
plt.show()
```

Comparing background and signal datasets, we can tell out of the 14 variables, phi and t32 are two that doesn't clearly separate the background from signal

And now let's look at the variable 'pt', which is transverse momentum. It't the components of momentum of the collision products in the plane that's perpendicular to the axis of the beam of particles, we expect that most momentum in the colliding particles will cancel each other.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('pt'), bins = 1000)
a[0].set_title('Background transverse momentum')
a[0].set_ylabel('Counts')
a[0].set_xlabel('GeV/c')
a[1].hist(qcd_data.get('pt'), bins = 1000)
a[1].set_title('Signal data transverse momentum')
a[1].set_ylabel('Counts')
a[1].set_xlabel('GeV/c')

plt.show()
```

Note that the peak of the data is around 1115 GeV/c, as we are working in particle scale, the background data tends to be more gently peaked at different values. Therefore the signal follows a qualitatively different distribution.

And now we look at the pseudorapidity, which is the measure of the angle between the particle beam and the 3-momentum of the collision products.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('eta'), bins = 1000)
a[0].set_title('Background pseudorapidity')
a[0].set_ylabel('Counts')
a[1].hist(qcd_data.get('eta'), bins = 1000)
a[1].set_title('Signal data pseudorapidity')
a[1].set_ylabel('Counts')

plt.show()
```

We see that pseudorapidity is normally distributed about 0 in both datasets. However the signal data has less SD which means it's more tightly towards the middle.

And next we look at the azimuthal angle, which is the angle in the transverse plane between Pt and the horizontal.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('phi'), bins = 1000)
a[0].set_title('Background azimuthal angle')
a[0].set_ylabel('Counts')
a[0].set_xlabel('Radians')
a[1].hist(qcd_data.get('phi'), bins = 1000)
a[1].set_title('Signal data azimuthal angle')
a[1].set_ylabel('Counts')
a[1].set_xlabel('Radians')

plt.show()
```

Looks like it's pretty much evenly distributed for both datasets.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('mass'), bins = 1000)
a[0].set_title('Background invaiant mass')
a[0].set_ylabel('Counts')
a[0].set_xlabel('GeV/c^2')
a[1].hist(qcd_data.get('mass'), bins = 1000)
a[1].set_title('Signal data invariant mass')
a[1].set_ylabel('Counts')
a[1].set_xlabel('GeV/c^2')


plt.show()
```

Invariant mass associated to the jet is the frame independent quantity, which is a lorentz scalar. As we can see it's sharply peaked, and the predicted Higgs mass is around 125 GeV/(c^2), which is every close to the peak for both graphs.For the signal data, we see that the peaking is more extreme compared to the background.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('ee2'), bins = 1000)
a[0].set_title('Background two point ECF ratio')
a[0].set_ylabel('Counts')
a[1].hist(qcd_data.get('ee2'), bins = 1000)
a[1].set_title('Signal data two point ECF ratio')
a[1].set_ylabel('Counts')

plt.show()
```

This is the two point energy corrlation function ratio of the jet.The difference between background and the signal is pretty suprising: the background mean is close to 0, but has a long tail towards larger values, while the signal data is much more clustered, with a very small tail.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('ee3'), bins = 1000)
a[0].set_title('Background three point ECF ratio')
a[0].set_ylabel('Counts')
a[0].set_yscale('log')
a[1].hist(qcd_data.get('ee3'), bins = 1000)
a[1].set_title('Signal data three point ECF ratio')
a[1].set_ylabel('Counts')

plt.show()
```

And now we look at the ECF, we expect this to be more sharply peaked, and it turns out it is. If the lokelihood of finding two particles in a jet with uncorrelated energies is low, it would likely be even lower for three, as two of the particles are more likely to be adjacent. The background is both more sharply peaked and has longer tail than the signal. The signal has a very short tail. And we hardly see any notable distribution.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('d2'), bins = 1000)
a[0].set_title('Background three-to-two point ECF ratio')
a[0].set_ylabel('Counts')
a[0].set_yscale('log')
a[1].hist(qcd_data.get('d2'), bins = 1000)
a[1].set_title('Signal data three-to-two point ECF ratio')
a[1].set_ylabel('Counts')
a[1].set_yscale('log')

plt.show()
```

This is the ratio of ee3/ee2^3, just like the last variable the signal data has a smaller tail than the background and both right skewed.

```python
fig, a = plt.subplots(1, 2, figsize = (15, 8))

a[0].hist(higgs_data.get('angularity'), bins = 1000)
a[0].set_title('Background angularity')
a[0].set_ylabel('Counts')
a[1].hist(qcd_data.get('angularity'), bins = 1000)
a[1].set_title('Signal data angularity')
a[1].set_ylabel('Counts')

plt.show()
```

Angularity, we can see the signal is far more sharply peaked than the background, also the tail of the data is much less than the background as well.
